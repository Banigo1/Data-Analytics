    Mini Project 19 Student Information System: 

Build a simple student information system where:

•	A dictionary is used to store student information (e.g., name, age, grades).

•	Sets are used to track courses that students are enrolled in, ensuring no duplicates.

•	Implement features to add students, enroll them in courses, and retrieve their information.


```python
# Initialize the student information system
students = {}  # Dictionary to store student information

# Add a new student
student_id = 1
name = "Alice"
age = 20
if student_id not in students:
    students[student_id] = {
        'name': name,
        'age': age,
        'courses': set()  # Set to store courses
    }
    print(f"Student {name} added successfully.")
else:
    print("Student ID already exists.")

# Add another student
student_id = 2
name = "Bob"
age = 22
if student_id not in students:
    students[student_id] = {
        'name': name,
        'age': age,
        'courses': set()
    }
    print(f"Student {name} added successfully.")
else:
    print("Student ID already exists.")

# Enroll a student in a course
student_id = 1
course_name = "Mathematics"
if student_id in students:
    students[student_id]['courses'].add(course_name)
    print(f"Student {students[student_id]['name']} enrolled in {course_name}.")
else:
    print("Student ID not found.")

# Enroll the same student in another course
course_name = "Physics"
if student_id in students:
    students[student_id]['courses'].add(course_name)
    print(f"Student {students[student_id]['name']} enrolled in {course_name}.")
else:
    print("Student ID not found.")

# Enroll another student in a course
student_id = 2
course_name = "Chemistry"
if student_id in students:
    students[student_id]['courses'].add(course_name)
    print(f"Student {students[student_id]['name']} enrolled in {course_name}.")
else:
    print("Student ID not found.")

# Retrieve and display student information
student_id = 1
if student_id in students:
    student_info = students[student_id]
    print(f"Student ID: {student_id}")
    print(f"Name: {student_info['name']}")
    print(f"Age: {student_info['age']}")
    print(f"Courses Enrolled: {', '.join(student_info['courses']) if student_info['courses'] else 'No courses enrolled.'}")
else:
    print("Student ID not found.")

# Retrieve information for another student
student_id = 2
if student_id in students:
    student_info = students[student_id]
    print(f"Student ID: {student_id}")
    print(f"Name: {student_info['name']}")
    print(f"Age: {student_info['age']}")
    print(f"Courses Enrolled: {', '.join(student_info['courses']) if student_info['courses'] else 'No courses enrolled.'}")
else:
    print("Student ID not found.")
```

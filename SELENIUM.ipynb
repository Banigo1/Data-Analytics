```python
import pandas as pd
from selenium import webdriver
#from selenium.webdriver.support.ui import webDriverwait
from selenium.webdriver.chrome.service import Service
#selenium.webdriver.chrome.service import Service
from bs4 import BeautifulSoup
from time import sleep

data = []
chrome_option = webdriver.ChromeOptions()
page = 0

def getsite(url):

    # Setup selenium webdriver
    service = Service(executable_path="C:/Downloads/chromedriver-win64/chromedriver-win64/chromedriver.exe")
    
    driver = webdriver.Chrome(service=service)


    driver.get(url)
    sleep(15)
    html = driver.page_source
    page_soup = BeautifulSoup(html, "lxml")
    containers = page_soup.findAll('li',{'class':'bbe45_3oExY _22339_3gQb9'})
    
    for i in containers:
        name = i.find('h3')
        price = i.find("span",{"class":"d7c0f_sJAqi"})
        former_price = i.find("span",{"class":"f6eb3_1MyTu"})
        discount_price = i.find("span",{"class":"false _6c244_q2qap _6977e_X5mZi"})
        rating= i.find("div",{"class":"a455c_3G0na"})
        if name != None:
            product_name = name.text.strip() if name else 'N/A'
            product_price = price.text.strip() if price else 'N/A'
            product_former_price = former_price.text.strip() if former_price else 'N/A'
            product_discount_price = discount_price.text.strip() if discount_price else 'N/A'
            product_rating = rating.text.strip() if rating else 'N/A'

            product = {
                'name': f"{product_name}",
                'price': f"{product_price}",
                'former_price': f"{product_former_price}",
                'discount_price': f"{product_discount_price}",
                'rating': f"{product_rating}"
            }
            data.append(product)
            #print()
            #print(header.text.slit())
            #num = num + 1

    driver.close()
urll = 'https://www.konga.com/category/'
category = [
    'accessories-computing-5227', 
    'phones-tablets-5294', 
    'electronics-5261', 
    'konga-fashion-1259', 
    'home-kitchen-602', 
    'baby-kids-toys-8', 
    'beauty-health-personal-care-4', 
    'drinks-groceries-581'
]
# Initialize page variable

for cat in category:
    for ur in range(5, 10):
        url = f"{urll}{cat}?page={ur}"
        getsite(url)
    print('End of pages for category:', cat)

print('Finished processing all categories page.', ur)
```
